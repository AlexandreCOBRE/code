# Step 2: Import the Pandas library to help manipulate the datasets

```python
# Step 2: Import the Pandas library to help manipulate the datasets
import pandas as pd
df1 = pd.read_csv("Relatorio_UFPR_Jan-21.csv")
display (df1)
```

```python
# Step 2: Import the Pandas library to help manipulate the datasets
import pandas as pd
df2 = pd.read_csv("Relatorio_UFPR_Mar-21.csv")
display (df2)
```

```python
# Step 2: Import the Pandas library to help manipulate the datasets
import pandas as pd
df3 = pd.read_csv("Relatorio_UFPR_Abr-21.csv")
display (df3)
```

```python
# Step 2: Import the Pandas library to help manipulate the datasets
import pandas as pd
df4 = pd.read_csv("Relatorio_UFPR_Mai-21.csv")
display (df4)
```

```python
# Step 2: Import the Pandas library to help manipulate the datasets
import pandas as pd
df5 = pd.read_csv("Relatorio_UFPR_Jun-21.csv")
display (df5)
```

# Step 3: Merging the five different databases into a single dataset

```python
# Step 3: Merging the five different databases into a single dataset
df_merge1 = df1.append(df2)
df_merge2 = df_merge1.append(df3)
df_merge3 = df_merge2.append(df4)
df_final = df_merge3.append(df5)
```

```python
df_final
```

# Step 4: removing completely missing columns

```python
# Step 4: removing completely missing columns
df_final = df_final.drop("cidade em que foi atendido", axis = 1)
df_final = df_final.drop("paciente com sintomas?", axis = 1)
df_final = df_final.drop("paciente com calafrios?", axis = 1)
df_final = df_final.drop("paciente com dor de cabeça?", axis = 1)
df_final = df_final.drop("paciente com distúrbios olfativos?", axis = 1)
df_final = df_final.drop("paciente com distúrbios gustativos?", axis = 1)
df_final = df_final.drop("paciente com dispineia ou desconforto respiratório?", axis = 1)
df_final = df_final.drop("paciente com cianose?", axis = 1)
df_final = df_final.drop("paciente com saturação de oxigenio baixa?", axis = 1)
df_final = df_final.drop("pas", axis = 1)
df_final = df_final.drop("pad", axis = 1)
```

```python
# Step 5: Downloading the combined datasets
df_final.to_csv('combined_datasets.csv', index=False)
```
