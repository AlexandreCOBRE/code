# Step 2: Import the Pandas library to help manipulate the database

```python
# Step 2: Import the Pandas library to help manipulate the database
import pandas as pd
df1 = pd.read_csv("combined_replaced_20features_replec2.csv")
display (df1)
```

# Step 3: Remove unnecessary columns

```python
# Step 3: Remove unnecessary columns
df2 = df1.drop("estado do atendimento", axis = 1)
df2 = df2.drop("data de inÍcio dos sintomas", axis = 1)
df2 = df2.drop("resultado igg", axis = 1)
df2 = df2.drop("resultado igm", axis = 1)
df2 = df2.drop("teste covid anticorpo", axis = 1)
df2 = df2.drop("fabricante.1", axis = 1)
df2 = df2.drop("data do atendimento", axis = 1)
df2 = df2.drop("fabricante", axis = 1)
```

```python
df2
```

```python
# Selecting only those who took the antigen test

df_antigen = df2.loc[df2["teste covid antígeno"]==1]
```

```python
df_antigen
```

# Step 4: Filtering only patients who have been tested for COVID-19 by the antigen method

```python
# Step 4: Filtering only patients who have been tested for COVID-19 by the antigen method

df_pos = df_antigen.loc[df_antigen["resultado do antígeno"]==1]
```

```python
df_pos
```

```python
# Selecting only those who had a negative antigen test
df_neg = df_antigen.loc[df_antigen["resultado do antígeno"]==0]
```

```python
df_neg
```

```python
# Merging the two datasets (negative and positive)
df_merge = df_pos.append(df_neg)
```

```python
df_merge
```

```python
df_merge.info()
```

```python
# Converting the antigen result variable to int
df_merge['resultado do antígeno'] = df_merge['resultado do antígeno'].astype(int)
```

```python
df_merge.info()
```
