# Step 2: Import the Pandas library to help manipulate the databa

```python
# Step 2: Import the Pandas library to help manipulate the databa
import pandas as pd
df1 = pd.read_csv("combined_replaced_20features_replec2.csv")
display (df1)
```

# Step 3: Remove unnecessary column

```python
# Step 3: Remove unnecessary column
df2 = df1.drop("estado do atendimento", axis = 1)
df2 = df2.drop("data de inÍcio dos sintomas", axis = 1)
df2 = df2.drop("teste covid anticorpo", axis = 1)
df2 = df2.drop("fabricante.1", axis = 1)
df2 = df2.drop("data do atendimento", axis = 1)
df2 = df2.drop("fabricante", axis = 1)
df2 = df2.drop("resultado do antígeno", axis = 1)
df2 = df2.drop("resultado igg", axis = 1)
```

```python
df2
```

```python
# Selecting only those who had a positive IgM test

df_pos = df2.loc[df2["resultado igm"]==1]
```

```python
df_pos
```

# Step 4: Filtering only patients who have been tested for COVID-19 by the IgM method

```python
# Selecting only those who had a positive IgM test
df_neg = df2.loc[df2["resultado igm"]==0]
```

```python
df_neg
```

```python
# Merging both datasets (negative and positive)
df_merge = df_pos.append(df_neg)
```

```python
df_merge
```

```python
df_merge.info()
```

```python
# Converting the result variable to the IgM result
df_merge['resultado igm'] = df_merge['resultado igm'].astype(int)
```

```python
df_merge.info()
```

```python
# Downloading the combined datasets
df_merge.to_csv('igm.csv', index=False)
```

# Step 5: Installing and importing the Pycaret library (helps to perform Machine Learning)

```python
# Step 5: Installing and importing the Pycaret library (helps to perform Auto-Machine Learn)
!pip install Pycaret
from pycaret import classification
```

# Step 6: Pre-processing the data

```python
# Step 6: Pre-processing the data
classification_setup = classification.setup(data = df_merge, target = "resultado igm")
```

# Step 7: Training, validating and comparing the different models

```python
# Step 7: Training, validating and comparing the different models
modelos = classification.compare_models()
```
