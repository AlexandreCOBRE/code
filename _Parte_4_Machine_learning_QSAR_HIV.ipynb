## **Passo 1: importar o dataset dos descritores (variáveis independentes)**

```python
## 1.1. Visualizando o dataset importado

import pandas as pd
df1 = pd.read_csv("pubchem_descritores_3classes.csv")
display (df1.head())
```

##**Passo 2: Tratar o dataset dos descritores**

```python
## 2.1. Eliminando variáveis não informativas: "Unnamed: 0" e "Name"
df1 = df1.drop("Unnamed: 0", axis = 1)
df1 = df1.drop("Name", axis = 1)
```

```python
## 2.2. Visualizando o dataset dos descritores após a remoção das variáveis não informativas
df1
```

##**Passo 3: importar o dataset dos valores de "pIC50" (variável dependente)**

```python
## 3.1: Importando o dataset contendo os valores de pI50 (variável pendentes)
from google.colab import files
uploaded = files.upload()
```

```python
## 3.2. Visualizando os dados da variável dependente
df2 = pd.read_csv("dataset_3classes.csv")
display(df2)
```

##**Passo 4: Tratar o dataset da variável dependente**

```python
## 4.1. Selecionando apenas a variável dependente 
df2 = df2["pIC50"]
```

```python
## 4.2. Visualizando a variável dependente
display(df2)
```

## **Passo 5: Combinar os dois datasets (variáveis independentes e dependente)**

```python
## 5.1. Observando se o número de linhas (amostras) de ambas variáveis são iguais
df1.shape, df2.shape
```

```python
## 5.2. Juntando as variáveis independentes e dependentes
```

```python
df3 = pd.concat([df1, df2], axis = 1)
```

```python
## 5.3. Visualizando os dados conbinados
df3
```
